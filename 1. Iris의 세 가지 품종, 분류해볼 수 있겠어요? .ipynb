#### describe 표의 의미?
- 평균(mean)
- 표준편차(std)
- 최소값(min)
- 최대값(max)
- 중간값(50%)

- 3사분위수(75%)

#### 사분위수란?

- 2사분위수(중간값) = 데이터를 일렬로 늘어놓았을 때 정중앙의 값
- 만약 데이터 개수가 짝수라서 중앙값을 선택할 수 없다면 가운데 2개의 값의 평균을 사용함

### (3) train, test 데이터 분리
<hr/>

```python
x_train, x_test, y_train, y_test = train_test_split(digits_data, digits_label, test_size = 0.2, random_state = 7)
```

### (4) 다양한 모델로 학습시켜보기
<hr/>

### 가설

손글씨 데이터를 확인해봤을 때, 이 데이터는 손글씨 이미지를 10가지의 특성 데이터 (1~ 10)중 하나로 분류하는 데이터이다. 다중 분류에 해당하며 랜덤 포레스트와 로지스틱 회귀 모델 성능이 뛰어날 것으로 예상한다.<br/>
손글씨 데이터는 손글씨를 분류하는데에 있어 precision과 recall의 중요도를 따지지 않기 때문에, 평가 지표는 fi-score와 acuuracy로 생각할 수 있다.<br/>
이 때, 손글씨 데이터의 label값이 전체적으로 균등하므로, 평가지표를 <b style="color: red;">accuracy</b>로 보기로 했다. 

### 1) Decision Tree 사용해 보기

#### 가설 및 기대효과

결정 트리 모델 특성의 중요도를 확인해볼 수 있다.<br/>
트리 그래프로 출력하여 눈으로 직접 어떻게 분류가 되는지 확인할 수 있다.<br/>
결정 트리 모델의 결과를 바탕으로, 랜덤 포레스트와 비교한 후 랜덤 포레스트의 장점을 알 수 있다.

#### 실험해보기

```python
# 변수에 모델을 저장 / 랜덤 시드 지정하기
decision_tree = DecisionTreeClassifier(random_state = 32)
```

```python
# 모델 학습 및 예측하기
decision_tree.fit(x_train, y_train)
tree_y_pred = decision_tree.predict(x_test)
```

```python
# 실제 값과 비교하기
tree_y_pred
```

```python
y_test
```

```python
# 모델의 정확도 평가하기
accuracy_score(y_test, tree_y_pred)
```

```python
# 모델의오차행렬 확인해보기
confusion_matrix(y_test, tree_y_pred)
```

```python
# 결과의 Precision, Recall, F1 score 출력하기
print(classification_report(y_test, tree_y_pred))
```

```python
# 결정트리 모델의 클래스에 맞게 색을 칠하여 그림으로 모델 출력해보기 
plt.figure(figsize = (30, 30))
plot_tree(decision_tree, filled = True, class_names = True)
plt.show()
```

```python
# 특성의 중요도 출력하기
decision_tree.feature_importances_
```

#### 결과

결정 트리 모델을 적용했을 때, 모델의 정확도 평가는 <b style="color: red;">86%</b>로 아쉬운 성공률을 보였다<br/>
또한 특성의 중요도를 출력해보았을 때, <b style="color: blue;">여러개의 특성들이 클래스를 분류하는데에 있어 영향을 미치지 않은 것으로 확인되었다.</b><br/>
때문에 랜덤 포레스트를 적용해보았을 때 특성의 중요도가 어떻게 바뀌는지 확인해보고, 결과가 어떻게 달라졌는지 비교를 하고자 한다.

### 2) Random Forest 사용해 보기

#### 가설 및 기대효과

결정 트리 모델은 결정경계(decision boundary)가 데이터 축에 수직이어서 특정 데이터에만 잘 작동할 가능성이 높다.<br/>그렇기 때문에 위의 모델에서 특성의 중요도가 0인 특성들이 있는 것을 확인할 수 있었다. <br/>
Random Forest를 이용한다면, 특정 데이터에만 작동하지 않고 모든 특성들에 대해 작동하게된다면 분류와 특성의 중요도 등 결과가 달라질 것으로 예측이 된다.

#### 실험해보기

```python
# 변수에 모델을 저장 / 랜덤 시드 지정하기 
random_forest = RandomForestClassifier(random_state = 32)
```

```python
# 모델 학습 및 예측하기
random_forest.fit(x_train, y_train)
random_forest_y_pred = random_forest.predict(x_test)
```

```python
# 실제 값과 비교하기
random_forest_y_pred
```

```python
y_test
```

```python
# 모델의 정확도 평가하기
accuracy_score(y_test, random_forest_y_pred)
```

```python
# 모델의오차행렬 확인해보기
confusion_matrix(y_test, random_forest_y_pred)
```

```python
# 결과의 Precision, Recall, F1 score 출력하기
print(classification_report(y_test, random_forest_y_pred))
```

```python
# 특성의 중요도 비교해보기
print(random_forest.feature_importances_)
```

```python
# 특성의 중요도 데이터 프레임으로 나타내보기
temp_df = pd.DataFrame(random_forest.feature_importances_)
temp_df.index = digits_df.columns[:-1]
temp_df.columns = ["RandomForest"]
temp_df.sort_values("RandomForest", ascending = False)
```

#### 결과
